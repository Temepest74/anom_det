1. In the first exercise you will use generate data() function from pyod.utils.data
to generate a 3-dimensional dataset with 300 training samples and 200 test
samples using a contamination of 0.15.


```python
from pyod.utils.data import generate_data
import numpy as np

CONTAMINATION = 0.15
SEED = 22
x_train, x_test, y_train, y_test = generate_data(
    n_train = 300, 
    n_test = 200, 
    contamination = CONTAMINATION,
    n_features = 3,
    random_state = SEED
)
```

2. You will fit an OCSVM model (from pyod.models.ocsvm) using the
training data and then you will predict the labels for test data. Use the
linear kernel and the contamination rate specified above. Compute the
balanced accuracy and the ROC AUC for the test data.

```python
from pyod.models.ocsvm import OCSVM;
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import roc_auc_score

model_ocsvm = OCSVM(
    kernel = "linear",
    contamination = CONTAMINATION
)

model_ocsvm.fit(x_train)
y_test_pred = model_ocsvm.predict(x_test)

print("BA:", balanced_accuracy_score(y_test, y_test_pred))
print("ROC AUC:", roc_auc_score(y_test, y_test_pred))
```

EX 2

1. In this exercise you will use the cardio dataset from ODDS (https://
odds.cs.stonybrook.edu/cardiotocogrpahy-dataset/). You will split
the dataset into a training set consisting of 40% of the data and a testing
set.

```python
import scipy.io as sio
from sklearn.model_selection import train_test_split

odds = sio.loadmat('cardio.mat')
x = odds['X']
y = odds['y']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=SEED)
```

2. You will use a cross-validation grid-search over a predefined map of possible parameters of OneClassSVM (from sklearn.svm) in order to find
the ones that offer best results. The map of possible parameters must
contain different kernels, different values for gamma parameter (for kernels that use it) and different values for the ν parameter (including the
contamination rate of train set).


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import OneClassSVM

grid = GridSearchCV(
    estimator = OneClassSVM(),
    n_jobs = -1,
    # linear does not use gamma. if we put in the first one, the lin will still try with each gamma
    param_grid = [
    {
        'kernel': ['poly', 'rbf', 'sigmoid'],
        'gamma': [0.001, 0.01, 0.1, 1],
        'nu': np.arange(0.1, 1, 0.2)
    },
    {
        'kernel': ['linear'],
        'nu': np.arange(0.1, 1, 0.1)
    }],
    scoring = 'accuracy',
)

grid.fit(x_train, y_train)
print(grid.best_estimator_.get_params())
```

3. GridSearchCV from sklearn.model selection will be used to perform the
grid-search over the training set using Balanced Accuracy as score. The
same metric will be reported for the model trained with the best parameters discovered by GridSearchCV (tested on the test set).


```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import OneClassSVM

grid = GridSearchCV(
    estimator = OneClassSVM(),
    n_jobs = -1,
    # linear does not use gamma. if we put in the first one, the lin will still try with each gamma
    param_grid = [
    {
        'kernel': ['poly', 'rbf', 'sigmoid'],
        'gamma': [0.001, 0.01, 0.1, 1],
        'nu': np.arange(0.1, 1, 0.1)
    },
    {
        'kernel': ['linear'],
        'nu': np.arange(0.1, 1, 0.1)
    }],
    scoring = 'balanced_accuracy',
)

grid.fit(x_train, y_train)

best_params = grid.best_params_
best_model = OneClassSVM(**best_params)
best_model.fit(x_train, y_train)

print("BA:", balanced_accuracy_score(y_test, best_model.predict(x_test)))
```

4. GridSearchCV will have to perform standardization; for this, it will
use a Pipeline from sklearn.pipeline that will perform standardization
using StandardScaler before fitting the OneClassSVM model.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

estimator = Pipeline([
    ('scale', StandardScaler()),
    ('ocSVM', OneClassSVM())
])

grid = GridSearchCV(
    estimator = estimator,
    n_jobs = -1,
    # linear does not use gamma. if we put in the first one, the lin will still try with each gamma
    param_grid = [
    {
        'ocSVM__kernel': ['poly', 'rbf', 'sigmoid'],
        'ocSVM__gamma': [0.001, 0.01, 0.1, 1],
        'ocSVM__nu': np.arange(0.1, 1, 0.1)
    },
    {
        'ocSVM__kernel': ['linear'],
        'ocSVM__nu': np.arange(0.1, 1, 0.1)
    }],
    scoring = 'balanced_accuracy',
)

grid.fit(x_train, y_train)

# Clean the parameter names
best_params = {
    key.replace('ocSVM__', ''): value 
    for key, value in grid.best_params_.items()
}

best_model = OneClassSVM(**best_params)
best_model.fit(x_train, y_train)

print("BA:", balanced_accuracy_score(y_test, best_model.predict(x_test)))
```

5. You have to pay attention to the labels returned by OneClassSVM model
(from sklearn): -1 for outliers and 1 for inliers. If you want to convert
them to the pyod format you can use
label pyod = (−1 ∗ label sklearn + 1)/2
In the current exercise you will perform the conversion from pyod format
to sklearn for the ground truth labels

```python
def convert_to_sklearn_labels(y):
    return np.where(y == 0, -1, 1)

y_train_converted = convert_to_sklearn_labels(y_train)
y_test_converted = convert_to_sklearn_labels(y_test)

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

estimator = Pipeline([
    ('scale', StandardScaler()),
    ('ocSVM', OneClassSVM())
])

grid = GridSearchCV(
    estimator = estimator,
    n_jobs = -1,
    # linear does not use gamma. if we put in the first one, the lin will still try with each gamma
    param_grid = [
    {
        'ocSVM__kernel': ['poly', 'rbf', 'sigmoid'],
        'ocSVM__gamma': [0.001, 0.01, 0.1, 1],
        'ocSVM__nu': np.arange(0.1, 1, 0.1)
    },
    {
        'ocSVM__kernel': ['linear'],
        'ocSVM__nu': np.arange(0.1, 1, 0.1)
    }],
    scoring = 'balanced_accuracy',
)

grid.fit(x_train, y_train_converted)

# Clean the parameter names
best_params = {
    key.replace('ocSVM__', ''): value 
    for key, value in grid.best_params_.items()
}

best_model = OneClassSVM(**best_params)
best_model.fit(x_train, y_train_converted)

print("BA:", balanced_accuracy_score(y_test_converted, best_model.predict(x_test)))
```

1. For this exercise we will need to download the shuttle dataset from ODDS
(https://odds.cs.stonybrook.edu/shuttle-dataset/). Load the data
using scipy.io.loadmat() and use train test split() to split it into train
and test subsets (use 50% of data for testing). Normalize your data accordingly.


```python
from scipy.io import loadmat
from sklearn.preprocessing import Normalizer

data = loadmat("shuttle.mat")
x = data['X'] 
y = data['y'].ravel() 

x_train, x_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.50, 
    random_state=SEED  
)

scaler = Normalizer()

scaler.fit(x_train)
x_train_normalized = scaler.transform(x_train)
x_test_normalized = scaler.transform(x_test)
```

2. Fit OCSVM and DeepSVDD models (from pyod) using the training data and compute the balanced accuracy (BA) and the area under the curve for the testing data for both models (ROC AUC - using
sklearn.metrics.roc auc score).


```python
model_ocsvm = OCSVM()
model_deep = DeepSVDD(verbose=0)

model_ocsvm.fit(x_train_normalized)
model_deep.fit(x_train_normalized)

print("BA OSCVM:", balanced_accuracy_score(y_train, model_ocsvm.labels_))
print("BA DeepSVDD:", balanced_accuracy_score(y_train, model_deep.labels_))

print("ROC AUC OSCVM:", roc_auc_score(y_test, model_ocsvm.predict(x_test_normalized)))
print("ROC AUC DeepSVDD:", roc_auc_score(y_test, model_deep.predict(x_test_normalized)))
```

3. Use different architectures for DeepSVDD and compute the 2 aforementioned metrics for each case.


```python
model_deep_1 = DeepSVDD(
    hidden_neurons=[64, 32, 16, 8, 4],
    verbose=0
)
model_deep_1.fit(x_train_normalized)
print("BA DeepSVDD 1:", balanced_accuracy_score(y_train, model_deep_1.labels_))
print("ROC AUC DeepSVDD 1:", roc_auc_score(y_test, model_deep_1.predict(x_test_normalized)))

model_deep_2 = DeepSVDD(
    hidden_neurons=[64, 64, 64, 64, 64],
    verbose=0
)
model_deep_2.fit(x_train_normalized)
print("BA DeepSVDD 2:", balanced_accuracy_score(y_train, model_deep_2.labels_))
print("ROC AUC DeepSVDD 2:", roc_auc_score(y_test, model_deep_2.predict(x_test_normalized)))

model_deep_3 = DeepSVDD(
    hidden_neurons=[256, 128, 64, 32, 16, 8, 4],
    verbose=0
)
model_deep_3.fit(x_train_normalized)
print("BA DeepSVDD 3:", balanced_accuracy_score(y_train, model_deep_3.labels_))
print("ROC AUC DeepSVDD 3:", roc_auc_score(y_test, model_deep_3.predict(x_test_normalized)))

model_deep_4 = DeepSVDD(
    hidden_neurons=[256, 128, 64, 32, 16, 8, 4],
    hidden_activation='leaky_relu',
    output_activation='linear',
    epochs=200,
    dropout_rate=0.1, 
    verbose=0
)
model_deep_4.fit(x_train_normalized)
print("BA DeepSVDD 4:", balanced_accuracy_score(y_train, model_deep_4.labels_))
print("ROC AUC DeepSVDD 4:", roc_auc_score(y_test, model_deep_4.predict(x_test_normalized)))
```
